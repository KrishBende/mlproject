```python
numeric_features=[feature for feature in df.columns if df[feature].dtype!='O']
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']

print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
df['total_score']=df['math_score']+df['reading_score']+df['writing_score']
df['average']=df['total_score']/3
```

```python
df.head(3)
```

```python
mathfull=df[df['math_score']==100]['math_score'].count()
writingfull=df[df['writing_score']==100]['writing_score'].count()
readingfull=df[df['reading_score']==100]['reading_score'].count()

print(f"Number of Students with full in Math- {mathfull}")
print(f"Number of Students with full in Writing- {writingfull}")
print(f"Number of Students with full in Reading- {readingfull}")
```

```python
print("Math 25 Percentile Score- ", (df.describe()['math_score']['25%']))
print("Writing 25 Percentile Score- ", (df.describe()['writing_score']['25%']))
print("Reading 25 Percentile Score- ", (df.describe()['reading_score']['25%']))
print()

mathless=df[df['math_score']<(df.describe()['math_score']['25%'])]['math_score'].count()
writingless=df[df['writing_score']<(df.describe()['writing_score']['25%'])]['writing_score'].count()
readingless=df[df['reading_score']<(df.describe()['reading_score']['25%'])]['reading_score'].count()

print(f"Number of Students scoring less than 25 Percentile in Math- {mathless}")
print(f"Number of Students scoring less than 25 Percentile in Writing- {writingless}")
print(f"Number of Students scoring less than 25 Percentile in Reading- {readingless}")
```

```python

reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```
